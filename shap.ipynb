```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import shap
import lightgbm
import platform
platform.system()

# 운영체제별 한글 폰트 설정
if platform.system() == 'Darwin': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')

plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정

# 글씨 선명하게 출력하는 설정
%config InlineBackend.figure_format = 'retina'

std_info_grad_combined_eng = pd.read_csv('./data/eng_data/std_info_grad_combined_eng.csv')
std_info_grad_ms_eng = pd.read_csv('./data/eng_data/std_info_grad_ms_eng.csv')
std_info_grad_phd_eng = pd.read_csv('./data/eng_data/std_info_grad_phd_eng.csv')

std_info_grad_combined = pd.read_csv('./data/original_data/std_info_grad_combined.csv')
std_info_grad_ms = pd.read_csv('./data/original_data/std_info_grad_ms.csv')
std_info_grad_phd = pd.read_csv('./data/original_data/std_info_grad_phd.csv')

grad_cat_features = ['학생소속대학명', '학생교육부계열', '성별', '입학과정', '학과', '학부출신']    #범주형 데이터
grad_num_features = ['gpa', '서류점수', '면접점수', '월별인건비']       #연속형 데이터
grad_term = ['rec014_school_term']
```

```python
for i in grad_cat_features:
    std_info_grad_combined_eng[i] = std_info_grad_combined_eng[i].astype('category')
    
X = std_info_grad_combined_eng[grad_cat_features + grad_num_features]
y = std_info_grad_combined_eng[grad_term]

model = lightgbm.LGBMRegressor()
model.fit(X,y)

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)

shap.summary_plot(shap_values, X)
```
